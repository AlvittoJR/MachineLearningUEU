```python
# Import Library

import requests
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
# Unduh teks dari URL

url = "https://shared.djambred.my.id/uas-nim-ganjil.txt"
response = requests.get(url)
text = response.text
```

```python
# Hapus karakter non-alfanumerik dan ubah menjadi lowercase

import re
text = re.sub(r'\W', ' ', text)
text = text.lower()
```

```python
#Tokenisasi teks menjadi kata-kata

words = nltk.word_tokenize(text)
```

```python
# Unduh stopwords untuk Bahasa Indonesia

import nltk
nltk.download('stopwords')
```

```python
# Hapus stop words

stop_words = set(stopwords.words("indonesian"))
filtered_words = [word for word in words if word.lower() not in stop_words]
```

```python
# Buat WordCloud

wordcloud = WordCloud(width=800, height=400, max_words=200, background_color='white').generate(' '.join(filtered_words))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Hitung frekuensi kata-kata dan buat barplot

freq_dist = nltk.FreqDist(filtered_words)
freq_dist.plot(30, cumulative=False)
```

```python
# Analisis sentimen menggunakan TextBlob

blob = TextBlob(text)
sentiment = blob.sentiment
print("Sentimen: ", sentiment)
```
